**Separando a base em treino e teste (75% / 25%).**

```python
# Importando a biblioteca
from sklearn.model_selection import train_test_split
# Separando treino e teste
x_train, x_test, y_train, y_test = train_test_split(df_dummies.iloc[:,6:], df_dummies.iloc[:,:6], test_size=0.25, shuffle=True)
# Visualizando os tamanhos da base de treino e da base de teste
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

**Tratando os dados de ENTRADA da rede**


```python
# Normalizando os Dados do x_train e aplicando a métrica no x_test
from sklearn import preprocessing
preprocessParams = preprocessing.StandardScaler().fit(x_train)
x_train_normalized = preprocessParams.transform(x_train)
x_test_normalized = preprocessParams.transform(x_test)
# Visualizando um resultado pontual da normalização realizada
x_train_normalized[0:2,0:2]
```

```python
# Transformando o conjunto de saídas de TREINO em array
resultados_train = pd.DataFrame(y_train[['Coluna1','Coluna2','Coluna3','Coluna4','Coluna5','Coluna6']]).to_numpy()
y_train_final = resultados_train[0:]
# Verificando a transformação
y_train_final[0:]
```

```python
# Transformando o conjunto de saídas de TESTE em array
resultados_test = pd.DataFrame(y_test[['Coluna1','Coluna2','Coluna3','Coluna4','Coluna5','Coluna6']]).to_numpy()
y_test_final = resultados_test[0:]
# Verificando a transformação
y_test_final[0:1]
```

```python
# Definindo função que transformará os resultados em encode
def encode_sena(x):
  x_encoded = []
  for jogo in x:
    jogo = [value - 1 for value in jogo]
    encoded = np.zeros(60)
    encoded[jogo] = 1
    x_encoded.append(encoded.astype(int))
  return np.array(x_encoded)
```

```python
# Transformando o y_train e y_test no encode criado, com 60 saídas (que serão 60 neurônios de saída na rede a ser criada)
y_train_encoded = encode_sena(y_train_final)
y_test_encoded = encode_sena(y_test_final)
# Verificando o formato final alcançado
print(y_train_encoded[0:1])
print(y_test_encoded[0:1])
```

```python
# Verificando o formato final alcançado
print(y_train_encoded[0:1])
print(y_test_encoded[0:1])
```

# **3ª ETAPA: Rede Neural Multi-label**

```python
# Importando as bibliotecas
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
# Montando a rede multilabel, com entrada = ao tamanho de atributos e saída = à 60 neurônio (cada um correspondente a 1 possível dezena)
RN = Sequential()
RN.add(Dense(80,input_shape = x_train_normalized.shape[1:] ,activation = 'sigmoid'))
RN.add(Dense(500, activation = 'sigmoid'))
RN.add(Dense(300, activation = 'sigmoid'))
RN.add(Dense(500, activation = 'sigmoid'))
RN.add(Dense(100, activation = 'sigmoid'))
RN.add(Dense(60, activation='softmax'))
RN.summary()
```

```python
# Rodando a rede neural, escolhendo otimizador, loss, e métricas de verificação dos resultados alcançados
RN.compile(optimizer= 'sgd', loss="binary_crossentropy", metrics=["accuracy"])
H = RN.fit(x_train_normalized, y_train_encoded, batch_size=500, epochs=10000, verbose=2, validation_split=0.2)
```

```python
# Avaliando a rede: evolução de loss e acurária
import matplotlib.pyplot   as plt
# 1. Loss
plt.subplot(2, 1, 1)
plt.plot(H.history['loss'])
plt.title('loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss'], loc='upper left')
plt.show()
# 2. Acurácia
plt.subplot(2, 1, 2)
plt.plot(H.history['accuracy'])
plt.title('acc')
plt.ylabel('acc')
plt.xlabel('epoch')
plt.legend(['acc'], loc='upper left')
plt.show()
```

```python
# Avaliando a rede com os dados de teste: loss e acurácia do modelo
score = RN.evaluate(x_test_normalized, y_test_encoded, verbose = 0)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```

```python
# Gráfico do treinamento: custo do treino e validação
plt.plot(H.history['loss'], label='train')
plt.plot(H.history['val_loss'], label='validation')
plt.title('Loss de treino e validação')
plt.ylabel('loss')
plt.xlabel('época')
plt.legend();
```
